```python
import pandas as pd
import json

thresh = 0.41

prob_pred_df = pd.read_csv('/data/submisssions/mask_rcnn_furniture_0001_attempt4_raw.csv')

for i in range(len(prob_pred_df)):
    prob_pred = json.loads(prob_pred_df.iloc[i, 2])

    cls_dict = {}
    for cls, polygons in prob_pred['aabb'].items():
        pred_polygons = []
        if polygons:
            for pol, prob in polygons:
                if prob > thresh:
                    pred_polygons.append(pol)
        cls_dict[cls] = pred_polygons
    pred = {"aabb": cls_dict}
    prob_pred_df.iloc[i, 2] = json.dumps(pred)
    
```

```python
prob_pred_df.to_csv('/data/submisssions/mask_rcnn_furniture_0001_attempt4_%.2f.csv'%thresh)
```
